```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/cereal (1).csv')
data.head()
```

```python
avg_ratings = data.groupby('mfr')['rating'].mean().sort_values()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='mfr', y='rating', data=avg_ratings)
```

```python
plt.xticks(range(0,101,10))
avg_ratings.plot(kind='bar', color='skyblue')
plt.xlabel('Manufacturer')
plt.ylabel('Average Rating')
plt.title('Average Ratings of Cereals by Manufacturer')


plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='calories', y='rating', data=data)
```
